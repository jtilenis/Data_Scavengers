```python
import requests
import json
import requests
from pprint import pprint
import pandas as pd
import gmaps
import matplotlib.pyplot as plt
```

```python
# Store filepath in a variable
census_file = "Census_Data_v3.csv"
census_data = pd.read_csv(census_file, encoding="ISO-8859-1")
census_data
```

```python
# delete NaN records from Census Data
new_data1 = census_data.dropna()

# delete median income less than 0
new_data = new_data1.drop(new_data1[new_data1["Median Income"] < 0].index)
new_data
```

```python
# Get median income and sum the population for each city
median_income = new_data.groupby('City')["Median Income"].median()
print(f"median income {median_income}")
Total_Population = new_data.groupby('City')["Population"].sum()
print(f"Total Population {Total_Population}")
```

```python
Total_Population.sort_values()
```

```python
# Extract first record from each group
new_data = census_data.groupby('City').first()
city_census_data = new_data.drop(columns = ["Median Income","Population"])
city_census_data
```

```python
# Add Median Income and Population to city census data
city_census_data["Median Income"]=median_income
city_census_data["Total Population"] = Total_Population
city_census_data
```

```python
# Output to file
city_census_data.to_csv("Resources/Clean_Census_Data.csv", index=True, header=True)
```
